```python
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
```

Amostras aleatorias a cada iteração

```python
# Regression                  #Regressao linear simples == 1 categoria
x, y = make_regression(n_samples=200, n_features=1, noise=30)

plt.scatter(x, y)
```

```python
modelo = LinearRegression()
modelo.fit(x, y)
```

Lembrando da função que monta a reta:

y = a +bx

OU SEJA:

y = l_coeff + a_coeff * x

```python
a_coeff = modelo.coef_
l_coeff = modelo.intercept_
```

```python
print("Coeficiente Angular: {0}\nCoeficiente Linear: {1}".format(a_coeff, l_coeff))
```

```python
y_pred = l_coeff + a_coeff * x
print(y_pred)
```

```python
plt.scatter(x, y)
plt.plot(x, l_coeff + a_coeff * x, color="red")

#Exemplo de previsao, informando o ponto X = 2.5
plt.scatter(1.5, l_coeff + a_coeff * 2.5, color="orange", s=130) # S = tamanho = size
plt.show()
```
