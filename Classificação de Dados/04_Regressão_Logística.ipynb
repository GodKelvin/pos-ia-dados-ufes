# Introduo  Regresso Logstica

Na aula de hoje, veremos uma segunda maneira de fazer classificao de dados: a **Regresso Logstica**.

Considere o contexto dado a seguir.

Suponha que voc esteja observando uma varivel binria (i.e., que pode assumir dois valores, 0 ou 1).

Esses valores podem indicar Passar ou Reprovar, Admitido ou Rejeitado, Esquerda ou Direita, etc.

Suponha tambm que exista uma **probabilidade** de se observar o valor 1, e que essa probabilidade  uma funo de algumas variveis (caractersticas).

Assim, as propriedades-chave de um problema que o fazem apropriado para ser abordado com Regresso Logstica so:
- O que voc pode observar so variveis **categricas** (a classe ou rtulo);
- Voc tambm tem interesse em estimar a **probabilidade** de se observar um rtulo especfico.

Vamos elaborar mais esse contexto por meio de um exemplo.

## Exemplo - Probabilidade de ser admitido na Ps-Graduao

Exemplo baseado no material disponvel em: https://stats.oarc.ucla.edu/r/dae/logit-regression/

Suponha um pesquisador interessado em saber como variveis, como GRE (*Graduate Record Exam*), GPA (*Grade Point Average*) e o prestgio da Universidade onde o aluno fez graduao podem afetar a admisso do aluno na ps-graduao.

**Observao:** esse exemplo e os dados utilizados so baseados no contexto da ps-graduao nos Estados Unidos.

Assim, h trs variveis preditoras (caractersticas ou atributos): ``gre``, ``gpa`` e ``rank``.
- ``gre`` e ``gpa`` sero tratadas como contnuas.
- ``rank`` tem valores inteiros de 1 a 4. ``rank`` com valor 1 indica o prestgio mais alto, enquanto o valor 4 indica o prestgio mais baixo.

Cada linha da matriz abaixo  referente a uma aplicao para uma ps-graduao. A coluna ``admit`` indica se o aluno aplicante foi aceito ou no (1 para sim e 0 para no).

```python
import pandas as pd

df = pd.read_csv('https://stats.idre.ucla.edu/stat/data/binary.csv')
df.head()
```

Vejamos algumas caractersticas do conjunto de dados.

```python
#@title
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn import datasets
import seaborn as sns
sns.set()

import warnings
warnings.filterwarnings('ignore')

ax = plt.figure(figsize = (10,7)).add_subplot()
df.hist(ax = ax)
```

Vamos olhar tambm para a probabilidade de admisso em funo de cada uma das caractersticas contidas no conjunto de dados.

Primeiro, o ``rank``.

```python
ax = df.groupby('rank').mean()['admit'].plot(marker = 'o', fontsize = 12)
ax.set_ylabel('Prob. de Admisso', fontsize = 16)
ax.set_xlabel('Rank', fontsize = 16);
```

Agora, o ``gre``.

```python
ax = df.groupby('gre').mean()['admit'].plot(marker = 'o', fontsize = 12)
ax.set_ylabel('Prob. de Admisso', fontsize = 16)
ax.set_xlabel('GRE', fontsize = 16);
```

Por fim, o ``gpa`` (para facilitar a visualizao, os valores de ``gpa`` so agrupados em 10 intervalos).

```python
bins = np.linspace(df.gpa.min(), df.gpa.max(), 10)
ax = df.groupby(np.digitize(df.gpa, bins)).mean()['admit'].plot(marker = 'o', fontsize = 12)
ax.set_ylabel('Prob. de Admisso', fontsize = 16)
ax.set_xlabel('Intervalo de GPA', fontsize = 16);
```

Pode-se perceber que h uma boa correlao entre cada uma das trs caractersticas e a probabilidade de um aluno ser admitido na ps-graduao.

Se houver interesse em estimar a **probabilidade** de admisso de um candidato, a Regresso Logstica pode ser uma ferramenta interessante!

Alm disso, uma vez que a probabilidade for estimada, ela pode ser utilizada para classificao (Entre admisso ou no)! Veremos como isso pode ser feito mais adiante!

## Formulao bsica (verso simplificada)

Como mencionado anteriormente, a Regresso Logstica se preocupa em estimar a **probabilidades**! Mas o que est disponvel no conjunto de dados so observaes de uma varivel categrica (no exemplo, ser admitido ou no na ps-graduao).

Poderamos tentar estimar tais probabilidades fazendo uso de um modelo linear do tipo $\alpha + \mathbf{\beta}^T\mathbf{x}$, onde:
- $\mathbf{x} \in \mathbb{R}^d$ representa o vetor de caractersticas (no exemplo acima, ``gre``, ``gpa`` e ``rank``);
- $\mathbf{\beta} \in \mathbb{R}^d$  o vetor de coeficientes, o qual indica o "peso" de cada caracterstica de $\mathbf{x}$;
- $\mathbf{\beta}^T$ representa o vetor linha, trasposto do vetor $\mathbf{\beta}$;
- $\mathbf{\beta}^T\mathbf{x}$ representa o produto escalar entre $\mathbf{x}$ e $\mathbf{\beta}$, i.e., $\mathbf{\beta}^T\mathbf{x} = \sum_{i = 1}^d \beta_ix_i$; e
- $\alpha \in \mathbb{R}$  uma constante.

No entanto, uma funo linear pode assumir qualquer valor real, e probabilidades devem estar restritas ao intervalo $(0, 1)$.

Para resolver esse problema, a Regresso Logstica tenta estimar com uma funo linear uma transformao das probabilidades: os **log-odds** ou **logits**.

**Definio:** para qualquer probabilidade $p \in (0, 1)$, a **odd** associada a $p$  dada por $$\frac{p}{1-p}.$$




**Definio:** para qualquer probabilidade $p  \in (0, 1)$, o **log-odd** ou **logit** de $p$  dado por $$\log{\frac{p}{1-1}}.$$

Dessa forma, a Regresso Logstica tenta encontrar um modelo linear para prever os logits. Em outras palavras, a ideia  escrever

$$\alpha + \mathbf{\beta}^T\mathbf{x} = \log{\frac{p(\mathbf{x})}{1 - p(\mathbf{x})}},$$

onde $p(\mathbf{x})$ representa a probabilidade de um objeto $\mathbf{x}$ ser da classe 1 (no nosso exemplo, de ser admitido na ps-graduao).




Observe que a equao acima pode ser manipulada de forma que $p(\mathbf{x})$ possa ser obtida em funo de $\alpha$ e $\mathbf{\beta}$:

$$p(\mathbf{x}) = \frac{e^{\alpha + \mathbf{\beta}^T\mathbf{x}}}{1 + e^{\alpha + \mathbf{\beta}^T\mathbf{x}}}.$$




## Na Prtica

Em resumo, a entrada e a sada da Regresso Logstica so

### Entrada

Um conjunto de dados $D = \{(\mathbf{x}_i, y_i)\}_{i = 1}^n$, onde os $\mathbf{x}_i$'s descrevem os objetos de interesse (vetores de caractersticas) e os $y_i$'s so os rtulos (classes) de cada objeto.

### Sada

Uma funo $p$, que para um dado vetor de caractersticas $\mathbf{x}$, permite calcular $p(\mathbf{x})$, as probabilidades de $\mathbf{x}$ pertencer a cada uma das possveis classes.

Em cenrios nos quais h apenas duas classes, geralmente, $p(\mathbf{x})$ representa a probabilidade de $\mathbf{x}$ pertencer  classe 1.

**Observao:** est fora do escopo dessa disciplina apresentar como os parmetros $\mathbf{\alpha}$ e $\mathbf{\beta}$ so estimados. Os interessados podem consultar as referncias disponveis no fim desse material.

## Como fazer a classificao?

Bom, se a Regresso Logstica no ajuda a estimar a probabilidade de um $\mathbf{x}$ pertencer a uma certa classe, como isso nos ajuda na tarefa de classificao?

Interessantemente, a Regresso Logstica pode ser vista como uma ferramenta de classificao. Considere, por exemplo, o caso de um problema de classificao binria (apenas classe 0 ou classe 1).

Para um dado $\mathbf{x}$ a ser classificado, podemos calcular $p(\mathbf{x})$.

Assim, associamos $\mathbf{x}$  classe 1 caso $p(\mathbf{x}) > 0.5$, e  classe 0 caso contrrio!

## Regresso Logstica na ``scikit-learn``




A biblioteca ``scikit-learn`` possui uma implementao de um classificador baseado em [Regresso Logstica](https://scikit-learn.org/1.5/modules/generated/sklearn.linear_model.LogisticRegression.html), que funciona de forma anloga aos classificadores que vimos anteriormente.

Vamos demonstrar o uso dessa implementao para o conjunto de dados utilizado no exemplo do incio da aula.

```python
X = df[['gpa', 'gre', 'rank']]
y = df.admit
```

Lembrando que, a avaliao de um classificador nunca deve ser feita com o mesmo conjunto de dados que foi utilizado para treinar o classificador.

Assim, vamos reter 30% dos dados para realizao de testes.

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30)
```

Aps isso, precisamos construir o classificador e realizar o treinamento com os dados do conjunto de treinamento.

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train, y_train);
```

Aps o treinamento, o modelo pode ser utilizado para prever a classe dos elementos do conjunto de teste.

```python
y_pred = model.predict(X_test);
```

Para avaliar o modelo obtido, podemos calcular a taxa de acertos (ou acurcia) dos elementos do conjunto de teste.

```python
from sklearn.metrics import accuracy_score

acc = accuracy_score(y_test, y_pred)

print("Acurcia do Classificador = %f" % acc)
```

## Regresso Logstica na ``scikit-learn`` - Outro Exemplo

De forma similar ao que fizemos com o classificador KNN na aula passada, vamos ver como o classificador baseado em Regresso Logstica se comporta no conjunto de dados ``wine``.

```python
from sklearn import preprocessing
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

# obtendo os dados
wine = datasets.load_wine(return_X_y = False, as_frame = True)
X = wine.data
y = wine.target.values

# devidindo em treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30)

# Aplicando transformao por reescala
scaler = preprocessing.MinMaxScaler(feature_range = (0, 1))
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)

# Ajustando o modelo
model = LogisticRegression()
model.fit(X_train_scaled, y_train);

# Aplicando a mesma transformao nos dados de teste
X_test_scaled = scaler.transform(X_test)

# Realizando a previso para os dados de teste
y_pred = model.predict(X_test_scaled);

# Medindo a acurcia do modelo nos dados de teste
acc = accuracy_score(y_test, y_pred)
print("Acurcia do Classificador = %f" % acc)
```

Pode-se perceber que o classificador, assim como o KNN, teve um bom resultado para o conjunto de dados ``wine``.
